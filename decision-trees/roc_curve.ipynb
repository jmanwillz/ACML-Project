```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import auc, roc_curve
```

```python
df_log_reg = pd.read_csv("../dataset/logreg_roc_values.csv")
df_log_reg
```

```python
df_rf = pd.read_csv("../dataset/rf_roc_values.csv")
df_rf
```

```python
# Logistic Regression
fpr1, tpr1, _ = roc_curve(df_log_reg["y_test"], df_log_reg["y_prob"])
roc_auc1 = auc(fpr1, tpr1)

# Random Forest
fpr2, tpr2, _ = roc_curve(df_rf["y_test"], df_rf["y_prob"])
roc_auc2 = auc(fpr2, tpr2)

# ANN
```

```python
# Plot both ROC curves
plt.figure(figsize=(8, 6))

plt.plot(fpr1, tpr1, label=f"Logistic Regression (AUC = {roc_auc1:.2f})")
plt.plot(fpr2, tpr2, label=f"Random Forest (AUC = {roc_auc2:.2f})")

plt.plot([0, 1], [0, 1], "k--", label="Random Classifier")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curves")
plt.legend(loc="lower right")

plt.grid(True)
plt.tight_layout()
plt.show()
```
